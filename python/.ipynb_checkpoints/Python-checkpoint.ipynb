# Question 1: -

Write a program that takes a string as input, and counts the frequency of each word in the string, there might
be repeated characters in the string. Your task is to find the highest frequency and returns the length of the
highest-frequency word.

Note - You have to write at least 2 additional test cases in which your program will run successfully and provide
an explanation for the same.


Example input - string = “write write write all the number from from from 1 to 100”

Example output - 5

Explanation - From the given string we can note that the most frequent words are “write” and “from” and
the maximum value of both the values is “write” and its corresponding length is 5

```python
def maxfreqwordlen(string):
    string.lower()
    list_of_words = string.split(' ')
    unique_words =list(set(list_of_words))
    print('unique words', unique_words)
    count= []
    for word in unique_words:
        count.append(list_of_words.count(word))
    print('unique words counts', count)    
    maxi = max(count)
    indices = [i for i in range(len(count)) if count[i] == maxi]
    word_len= 0
    for index in indices:
        if len(unique_words[index])>word_len:
            word_len  = len(unique_words[index])
        
    return word_len

maxfreqwordlen('write write write all the number from from from 1 to 100')
```

# Question 2: -
    
Consider a string to be valid if all characters of the string appear the same number of times. It is also valid if
he can remove just one character at the index in the string, and the remaining characters will occur the same
number of times. Given a string, determine if it is valid. If so, return YES , otherwise return NO .

Note - You have to write at least 2 additional test cases in which your program will run successfully and provide
an explanation for the same.

Example input 1 - s = “abc”. This is a valid string because frequencies are { “a”: 1, “b”: 1, “c”: 1 }

Example output 1- YES

Example input 2 - s “abcc”. This string is not valid as we can remove only 1 occurrence of “c”. That leaves
character frequencies of { “a”: 1, “b”: 1 , “c”: 2 }

Example output 2 - NO

```python
def checkstring(string):
    chars = [*string]
    unique_char =list(set(chars))
    print('unique char', unique_char)
    count= []
    for char in unique_char:
        count.append(chars.count(char))
    if all(element == count[0] for element in count):
        return 'YES'
    else:
        no_of_counts = list(set(count))
        if len(no_of_counts)>2:
            return 'NO'
        elif 1 in no_of_counts:
            if count.count(1) == 1:
                return 'YES'
            else: return 'NO'
        else: return 'NO'

checkstring('abc')
```

```python
print(checkstring('abbcc'))
print(checkstring('aaabbcc'))
print(checkstring('abbccc'))
```

```python
# Plotting the distribution as a pie plot
distribution = EV_data.groupby(['City', 'Electric Vehicle Type']).size().reset_index(name='Count')
import math
plt.figure(figsize=(15, 15))
wedges, _, autotexts = plt.pie(
    distribution.Count,
    labels=distribution.City,
    autopct='%1.1f%%',
    startangle=90
)

# Set the count as ticks outside the pie plot
plt.gca().set_aspect("equal")
plt.gca().set_anchor('C')
plt.gca().set_xlim(-1.5, 1.5)
plt.gca().set_ylim(-1.5, 1.5)

for i, (wedge, autotext) in enumerate(zip(wedges, autotexts)):
    angle = (wedge.theta2 + wedge.theta1) / 2
    x = 1.25 * wedge.r * 0.5 * math.cos(angle * (3.14159 / 180))
    y = 1.25 * wedge.r * 0.5 * math.sin(angle * (3.14159 / 180))
    plt.text(x, y, f"{distribution.Count[i]}", ha='center', va='center')

plt.title('Distribution of Electric Vehicle Type by City')
plt.show()
```

# Question 10 -
Write a program to count the number of verbs, nouns, pronouns, and adjectives in a given particular phrase or
paragraph, and return their respective count as a dictionary.

Note -

1. Write code comments wherever required for code
 
2. You have to write at least 2 additional test cases in which your program will run successfully and provide

```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from collections import defaultdict

def count_pos_tags(text):
    # Tokenize the text into words
    words = word_tokenize(text)

    # Remove stopwords
    stop_words = set(stopwords.words('english'))
    filtered_words = [word for word in words if word.casefold() not in stop_words]

    # Lemmatize the words
    lemmatizer = WordNetLemmatizer()
    lemmatized_words = [lemmatizer.lemmatize(word.lower(), pos=get_wordnet_pos(tag)) for word, tag in nltk.pos_tag(filtered_words) if get_wordnet_pos(tag)]

    # Count the number of verbs, nouns, pronouns, and adjectives
    pos_counts = defaultdict(int)
    for word, tag in nltk.pos_tag(lemmatized_words):
        if tag.startswith('VB'):  # Verb
            pos_counts['verbs'] += 1
        elif tag.startswith('NN'):  # Noun
            pos_counts['nouns'] += 1
        elif tag.startswith('PRP'):  # Pronoun
            pos_counts['pronouns'] += 1
        elif tag.startswith('JJ'):  # Adjective
            pos_counts['adjectives'] += 1

    return pos_counts

# Helper function to map NLTK's POS tags to WordNet POS tags
def get_wordnet_pos(tag):
    if tag.startswith('J'):
        return 'a'  # Adjective
    elif tag.startswith('V'):
        return 'v'  # Verb
    elif tag.startswith('N'):
        return 'n'  # Noun
    elif tag.startswith('PRP'):
        return 'pron'  # Adverb
    else:
        return None

# Example usage
text = "The quick brown fox jumps over the lazy dog. It is a beautiful day."

pos_counts = count_pos_tags(text)
print(pos_counts)
```
